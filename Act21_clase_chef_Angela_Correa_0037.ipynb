```python
class Chef:
    # El método __init__ es donde definimos los atributos (datos)
    def __init__(self, nombre, especialidad, experiencia):
        self.nombre = nombre
        self.especialidad = especialidad
        self.experiencia = experiencia

    # Definimos las funciones (acciones)
    def preparar(self):
        print(f"{self.nombre} está preparando los ingredientes para un platillo de {self.especialidad}.")

    def cocinar(self):
        print(f"{self.nombre} está en los fogones. ¡Huele de maravilla!")

    def emplatar(self):
        print(f"{self.nombre}, con {self.experiencia} años de experiencia, ha servido un plato espectacular.")

# --- Creación de Objetos (Instancias) ---

# Creando el objeto chef1 con datos de ejemplo
chef1 = Chef("Ramón Valdés", "Cocina Mexicana", 15)

# Creando el objeto chef2 con datos de ejemplo
chef2 = Chef("Julia Child", "Repostería Francesa", 20)

# --- Ejecución de las funciones ---

print("--- Actividad del Chef 1 ---")
chef1.preparar()
chef1.cocinar()
chef1.emplatar()

print("\n--- Actividad del Chef 2 ---")
chef2.preparar()
chef2.cocinar()
chef2.emplatar()
print("Programa realizado por Angela Correa NL = 13")
```
